## Viewing and Manipulating the Customer_shopping_behavior dataset

```python
import pandas as pd

df = pd.read_csv(r'C:\Users\nicks\Downloads\Customer_shopping_behavior.csv',encoding='latin1')
```

```python
# retrieving first 5 rows of dataset
df.head()
```

```python
df.info
```

```python
df.describe(include='all')
```

```python
# investigating the amount of NULL values in every column
df.isnull().sum()
```

```python
# filling the NULL values in the 'Review Rating' column with the median rating
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
# verifying the modification worked
df.isnull().sum()
```

```python
# making all the letters in the column names lower case
df.columns = df.columns.str.lower()

# replacing any spaces within the hearrs with underscores
df.columns = df.columns.str.replace(' ','_')

# renaming the 'purchase_amount_(usd)'' column
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
# verifying the cleansing worked
df.columns
```

```python
# create a column called 'age_group' with labels
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
# verifying the column creation worked
df[['age','age_group']].head(10)
```

```python
# create column 'purchase_frequency_days'

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7.0,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping).astype(int)
```

```python
# verifying the column creation worked and viewing the first 10 rows
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
# making sure the values in these two columns are the same
df[['discount_applied','promo_code_used']].head(10)
```

```python
# a more convenient way of doing the previous step when there are many rows
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# dropping the 'promo_code_used' column
df = df.drop('promo_code_used', axis=1)
```

## How to import a dataframe into a SQL Server database table

```python
from sqlalchemy import create_engine
import urllib
```

```python
# Construct the connection string
# Replace 'ODBC Driver 17 for SQL Server' with the appropriate driver name if different
params = urllib.parse.quote_plus('DRIVER={SQL Server};' "SERVER={DESKTOP-1RUIP5R\SQLEXPRESS};" "DATABASE={customer_behavior};" "trusted_connection=yes")

# Create the SQLAlchemy engine
engine = create_engine(f"mssql+pyodbc:///?odbc_connect={params}")
```

```python
df.to_sql('customer_shopping_behavior', con=engine, if_exists='replace', index=False)

print("DataFrame successfully imported to SQL Server.")
```

## Business Analysis

```python
# Total revenue generated male vs female customer

query='''select gender, SUM(purchase_amount) AS revenue from [customer_behavior].[dbo].[customer_shopping_behavior] GROUP BY gender'''

cursor=conn.cursor()
cursor.execute(query)
for row in cursor:
    print(row)
```

```python
# customer that used a discount but still spent more than the average purchase amount

query='''select customer_id
from [customer_behavior].[dbo].[customer_shopping_behavior]
where purchase_amount > 
(select AVG(purchase_amount)
from [customer_behavior].[dbo].[customer_shopping_behavior])'''

cursor=conn.cursor()
cursor.execute(query)
for row in cursor:
    print(row)
```

```python
# 5 products with highest average review rating

query='''select top 5 customer_id, AVG(review_rating)
from [customer_behavior].[dbo].[customer_shopping_behavior]
GROUP BY customer_id
ORDER BY AVG(review_rating) DESC'''
```

```python
cursor=conn.cursor()
cursor.execute(query)
for row in cursor:
    print(row)
```

```python
# Comparing average purchase amounts between standard and express shipping

query='''select shipping_type, AVG(purchase_amount)
from [customer_behavior].[dbo].[customer_shopping_behavior]
GROUP BY shipping_type
HAVING shipping_type in ('Standard','Express')
ORDER BY AVG(purchase_amount);'''
```

```python
cursor=conn.cursor()
cursor.execute(query)
for row in cursor:
    print(row)
```

```python
# Do subscribed customers spend more? Compare average spend and total revenue between subscribers and non-subscribers

query='''select subscription_status, AVG(purchase_amount), SUM(purchase_amount)
from [customer_behavior].[dbo].[customer_shopping_behavior]
GROUP BY subscription_status'''
```

```python
cursor=conn.cursor()
cursor.execute(query)
for row in cursor:
    print(row)
```
